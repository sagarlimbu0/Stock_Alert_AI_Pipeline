```python
import pandas as pd
import numpy as np

# Web scrapping package
from bs4 import BeautifulSoup as bs
import requests
```

```python
## Target URL: yfinance
url= 'http://finance.yahoo.com/markets/stocks/losers/'
```

```python
response = requests.get(url, headers= {"User-Agent": "Mozilla/5.0"})
```

```python
soup= bs(response.text,  "html.parser")
```

```python
content_list= list(soup)
```

#### Symbol

```python
# Find all matching <tr> elements
rows = soup.find_all('tr', class_='row yf-ao6als')

# Extract symbol and price from each row
for row in rows:
    cells = row.find_all('td')
    if len(cells) >= 2:
        symbol = cells[0].text.strip()
        price = cells[1].text.strip()
        print(f"{symbol}: {price}")
```

#### Get all stock `metadata`

```python
from bs4 import BeautifulSoup

# html = '''<your HTML with multiple <tr> rows>'''  # Replace with your full HTML content

# soup = BeautifulSoup(html, 'html.parser')
rows = soup.find_all('tr', class_='row yf-ao6als')

stock_data = []

for row in rows:
    try:
        symbol = row.select_one('.symbol').text.strip()
        company_name = row.select_one('.companyName').text.strip()
        price = row.find('fin-streamer', {'data-field': 'regularMarketPrice'}).text.strip()
        change = row.find('fin-streamer', {'data-field': 'regularMarketChange'}).text.strip()
        change_percent = row.find('fin-streamer', {'data-field': 'regularMarketChangePercent'}).text.strip()
        volume = row.find('fin-streamer', {'data-field': 'regularMarketVolume'}).text.strip()
        avg_volume = row.find_all('td')[7].text.strip()
        market_cap = row.find('fin-streamer', {'data-field': 'marketCap'}).text.strip()
        fifty_two_week_change = row.find('fin-streamer', {'data-field': 'fiftyTwoWeekChangePercent'}).text.strip()

        stock_data.append({
            'Symbol': symbol,
            'Company Name': company_name,
            'Price': price,
            'Change': change,
            'Change Percent': change_percent,
            'Volume': volume,
            'Average Volume': avg_volume,
            'Market Cap': market_cap,
            '52W Change %': fifty_two_week_change
        })
    except Exception as e:
        print("Skipping row due to error:", e)
        continue

# Print results
for stock in stock_data:
    print(stock)
```
